## When should I use "inplace" parameter in pandas?

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.shape
```

```python
ufo.head()
```

```python
ufo.drop('City',axis=1).head()
```

```python
ufo.head()
```

```python
ufo.drop('Time',axis=1, inplace=True)
```

```python
ufo.head()
```

```python
ufo.dropna(how='any')
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.shape
```

```python
ufo.dropna(how='any' , inplace=True)
```

```python
ufo.shape
```

```python
ufo= ufo.set_index('State')
```

```python
ufo.tail()
```

```python
ufo.fillna(method='bfill').tail()
```

```python
ufo.fillna(method='ffill').tail()
```
